Crear Sesión Spark

```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
SpSession = SparkSession \
           .builder \
           .appName("Demo Spark") \
           .master("local[*]")\
           .getOrCreate()
```

```python
SpContext = SpSession.sparkContext
```

Cargar Datos

```python
data = SpSession.read.csv('cars.csv', header=True, sep=";")
data.show(5)
```

```python
data = data.filter(data.Car != "STRING")
data.show(5)
```

```python
data.write.csv("cars2.csv", header=True, sep=";", mode="overwrite")
```

```python
data = SpSession.read.csv('cars2.csv', header=True, sep=";" , inferSchema= True )
data.show(5)
```

# Selección de Columnas

```python
# Hay varias formas
# Método 1:
data.select(data.Car).show(truncate=False)
```

```python
# Método 2:
data.select(data["Car"]).show(truncate=False)
```

```python
# Método 3
from pyspark.sql.functions import col

data.select(col('Car')).show(truncate=False)
```

# Agregar Nuevas Columnas

```python
# Caso 1:
from pyspark.sql.functions import lit
data=data.withColumn("Firts_columns", lit(1))
data.show(5)
```

```python
data = data.withColumn('Second_column', lit(2)).withColumn('Third_column', lit(3))
data.show(5)
```

# Agrupación

```python
data.groupBy("Horsepower").count().show()
```

```python
data.groupBy("Origin").count().show()
```

# Eliminación de Columnas

```python
data=data.drop("Second_column","Third_column")
data.show(5)
```
